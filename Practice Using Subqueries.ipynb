#  Practice problems using subqueries

1.  Using `IN`, find the names of all people who **produced** a 2019 movie that _Scarlett Johansson_ was in.

(3 rows).

```sql
use [jack.rutherford];
select FirstName, LastName, FullName, Title from imdb.People
    join imdb.Principals on Principals.PersonID = People.PersonID
    join imdb.Movies on Principals.TitleID = Movies.TitleID
where Principals.Category = 'Producer'
and YearReleased = 2019
and Movies.TitleID in (
    select Movies.TitleID from imdb.People SJ
    join imdb.Principals on SJ.PersonID = Principals.PersonID
    join imdb.Movies on Principals.TitleID = Movies.TitleID
    where FirstName = 'Scarlett'
      and LastName = 'Johansson'
      and IsActor = 1
)
```

2.  Using `EXISTS`, find customers who have placed an order for a product whose name contains _socks_, and an order whose name contains _pants_.  The two products could have been part of the **same** order, but they also could be from **different** orders.

(23 rows)

```sql
use [SalesOrdersExample];
select * from Customers
    join Orders on Orders.CustomerID = Customers.CustomerID
where exists(
    select * from Products p
        join Order_Details od on od.OrderNumber = Orders.OrderNumber
        where p.ProductName like '%socks%' and p.ProductNumber = od.ProductNumber 
    or exists(
        select * from Products p
        join Order_Details od on od.OrderNumber = Orders.OrderNumber
        where p.ProductName like '%pants%' and p.ProductNumber = od.ProductNumber 
    )
)
```
